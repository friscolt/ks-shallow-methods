```python
import cv2
import os
from google.colab.patches import cv2_imshow
```

Functions

```python
def f_readfiles(main_root,fldr_img,fldr_gt):
  folder_img = main_root + fldr_img; print(folder_img)
  folder_gt = main_root + fldr_gt; print(folder_gt)
  
  images = []
  for filename in os.listdir(folder_img):
    img = cv2.imread(os.path.join(folder_img,filename))
    gt = cv2.imread(os.path.join(folder_gt,filename),cv2.IMREAD_UNCHANGED)

    cv2_imshow(img)
    cv2_imshow(gt)
    
    if img is not None:
      images.append(img)
  return images 
```
